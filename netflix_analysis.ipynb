## Data Understanding
The dataset contains information about Netflix content including movies and TV shows. it includes details like title, director , cast, country and release year. 

```python
df['director'].fillna('unkown',inplace=True)
df['cast'].fillna('unkown',inplace=True)
df['country'].fillna('unkown',inplace=True)

df.dropna(subset=['rating'], inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x= 'type', data=df)
plt.title("Movies vs TV Shows on Netflix")
plt.show()
```

## Insight
Netflix has a higher number of movies compared  to TV shows.

```python
top_countries = df['country'].value_counts().head(10)
top_countries.plot(kind='bar')
plt.title("Top 10 Countries Producing Netflix Content")
plt.show()
```

## Insight
#The United states Produces the highest amount of Netflix content, followed by other countries like India and the United Kingdom.

```python
df['release_year'].value_counts().sort_index().plot()
plt.title("Netflix Content Growth Over Years")
plt.xlabel("year")
plt.ylabel("number of shows")
plt.show()
```

## Insight
Netflix content has increased rapidly over the years, especially after 2015, showing significant growth in the platform's production.

```python
genres = df['listed_in'].str.split(',', expands=True).stack()
genres.value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genres")
plt.ylabel("Count")
plt.show()
```

## Insight
Drama and International Movies are among the most popular genres on Netflix, Indicating a strong audience prefrence for diverse and emotional content.
